```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['age'])
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
```

```python
sns.histplot(df["bmi"])
plt.title("Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Frequency")
```

```python
sns.histplot(df["children"])
plt.title("Distribution of Children")
plt.xlabel("Children")
plt.ylabel("Frequency")
```

```python
sns.histplot(df["expenses"])
plt.title("Distribution of Expenses")
plt.xlabel("Expenses")
plt.ylabel("Frequency")
```

```python
sns.distplot(df['age'])
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
```

```python
sns.distplot(df["bmi"])
plt.title("Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Frequency")
```

```python
sns.distplot(df["children"])
plt.title("Distribution of Children")
plt.xlabel("Children")
plt.ylabel("Frequency")
```

```python
sns.distplot(df["expenses"])
plt.title("Distribution of Expenses")
plt.xlabel("Expenses")
plt.ylabel("Frequency")
```

```python
sns.countplot(x='sex', data=df)
plt.title("Count of Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
```

```python
sns.countplot(x='smoker', data=df)
plt.title("Count of Smokers")
plt.xlabel("Smoker")
plt.ylabel("Count")
```

```python
sns.countplot(x='region', data=df)
plt.title("Count of Region")
plt.xlabel("Region")
plt.ylabel("Count")
```

```python
correlation = df.corr(numeric_only=True)

sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
```

```python
print("Average Expenses:")
print(df["expenses"].mean())
```

```python
print("Maximum Expenses:")
print(df["expenses"].max())
```

```python
print("Minimum Expenses:")
print(df["expenses"].min())
```

```python
print("Average Expenses by Smoker:")
print(df.groupby("smoker")["expenses"].mean())
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(x="sex", y="expenses", hue="smoker", data=df)

plt.title("Distribution of Expenses by Sex and Smoker")
plt.xlabel("Sex")
plt.ylabel("Expenses")
```
