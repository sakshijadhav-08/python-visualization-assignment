```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Load Dataset
ds = pd.read_csv("ford_car_dataset.csv")

# Clean Data
ds = ds.drop_duplicates()
ds = ds.dropna()

# Histograms
ds["price"].hist()
plt.show()

ds["mileage"].hist()
plt.show()

ds["year"].hist()
plt.show()

ds["engineSize"].hist()
plt.show()

ds["mpg"].hist()
plt.show()

# Count Plots
sns.countplot(x="fuelType", data=ds)
plt.show()

sns.countplot(x="transmission", data=ds)
plt.show()

plt.figure(figsize=(12,5))
sns.countplot(y="model", data=ds)
plt.show()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(ds.select_dtypes(include="number").corr(), annot=True, cmap="coolwarm")
plt.show()

# Input and Output Features
X = ds.drop("price", axis=1)
y = ds["price"]

# One-Hot Encoding
ds = pd.get_dummies(ds, columns=["fuelType", "transmission", "model"], drop_first=True)

# Feature Scaling
X = ds.drop("price", axis=1)

num_cols = X.select_dtypes(include="number").columns

scaler = StandardScaler()

X[num_cols] = scaler.fit_transform(X[num_cols])

X.head()
```
